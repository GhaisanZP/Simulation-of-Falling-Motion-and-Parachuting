```python
# Pendefinisian Konstanta
mass = 90 # Massa
acc_gravity = -9.8  # Konstanta gravitasi
position = 2108 # Posisi awal
velocity = 0 # Kecepatan awal
speed = [[],[],[]] # Vektor untuk menyimpan hasil simulasi laju dari ketiga variasi waktu
heightvec = [[],[],[]] # Vektor untuk menyimpan hasil simulasi posisi dari ketiga variasi waktu
dt = 0.01 # Selang waktu yang digunakan untuk simulasi
tvec = [[],[],[]] # Vektor untuk menyimpan waktu
import matplotlib.pyplot as plt # Library untuk plot grafik
waktu = [15,20,25]
weight = mass * acc_gravity # Menghitung gaya berat
t = 0
for i in range(3) :
  while position > 10**(-6) and t <= 120:
    if t >= waktu[i] :
        projected_area = 28
    else :
        projected_area = 0.4 # Membuat kondisional untuk nilai projected area yang bergantung pada waktu
    air_friction = -0.65 * projected_area * velocity * abs(velocity) # Menghitung gaya gesek udara
    total_force = weight + air_friction # Menghitung gaya total
    acceleration = total_force/mass # Menghitung percepatan
    change_in_velocity = acceleration # Memasukkan nilai perubahan kecepatan
    change_in_position = velocity # Memasukkan nilai perubahan posisi
    velocity = velocity + change_in_velocity * dt # Mengganti nilai kecepatan
    position = position + change_in_position * dt # Mengganti nilai posisi
    speed[i].append(abs(velocity)) # Memasukkan nilai laju ke vektor
    heightvec[i].append(position) # Memasukkan nilai posisi ke vektor
    t = t + dt # Melakukan iterasi untuk waktu
    tvec[i].append(t) # Memasukkan nilai waktu ke vektor
  velocity = 0
  position = 2108
  t = 0
  # Mereset semua nilai
# Proses looping untuk menyimpan nilai yang diperlukan ke dalam

FinalPlot, Axes = plt.subplots(nrows=2, ncols=1, sharex=True, sharey=False)
FinalPlot.suptitle("Grafik Fungsi Posisi dan Laju Penerjun")
Axes[0].plot(tvec[0], heightvec[0], 'b', label="tinggi untuk t = 15 s")
Axes[0].plot(tvec[1], heightvec[1], 'y', label="tinggi untuk t = 20 s")
Axes[0].plot(tvec[2], heightvec[2], 'g', label="tinggi untuk t = 25 s")
Axes[0].legend()
Axes[0].grid()
Axes[0].set_xlabel("Waktu (s)")
Axes[0].set_ylabel("Tinggi (m)")

Axes[1].plot(tvec[0], speed[0], 'b', label="laju untuk t = 15 s")
Axes[1].plot(tvec[1], speed[1], 'y', label="laju untuk t = 20 s")
Axes[1].plot(tvec[2], speed[2], 'g', label="laju untuk t = 25 s")
Axes[1].legend()
Axes[1].grid()
Axes[1].set_xlabel("Waktu (s)")
Axes[1].set_ylabel("Laju (m/s)")

plt.show()
```
